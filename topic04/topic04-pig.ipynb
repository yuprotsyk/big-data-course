### 1. Що таке Apache Pig



**Pig** – це платформа для аналізу великих колекцій даних, що складається з:

- Мови високого рівня для написання програм аналізу

- Інфраструктури для запуску цих програм

### Pig

- Top Level [Apache Project](http://pig.apache.org)
    
- Pig - це високорівнева платформа поверх Hadoop
    - Надає мову програмування високого рівня **Pig Latin** для обробки даних
    
    - Перетворює код такої програми на MapReduce-задачі та виконує їх на кластері Hadoop

- Використовується у багатьох компаніях

### 2. Pig та MapReduce

- Для написання задач MapReduce потрібні програмісти
    - Які повинні вміти думати у стилі "map & reduce"
    
    - Швидше за все повинні знати мову Java

- Pig надає мову, яку можуть використовувати
    - Аналітики
    
    - Data Scientists
    
    - Статистики

- Спочатку був розроблений в компанії Yahoo! у 2006 для надання аналітикам доступу до даних

### 4. Компоненти Pig

- Pig Latin
    - Мова, заснована на командах

    - Розроблена спеціально для трансформації даних та послідовної обробки

- Середовище виконання
    - Середовище, в якому виконуються команди Pig Latin

    - Підтримка режимів виконання *Local* та *Hadoop*
    
- Компілятор Pig перетворює код Pig Latin на MapReduce
    - Компілятор оптимізує процес виконання

### 5. Режими виконання

- Local
    - Запускається в рамках однієї JVM
    
    - Працює виключно з локальною файловою системою

    - Відмінно підходить для розробки, експериментів та прототипів

    - `$pig -x local`
    
- Hadoop
    - Також відомий як режим MapReduce

    - Pig перетворює програму Pig Latin на MapReduce-задачі та виконує їх на кластері

    - `$pig -x mapreduce`

### 6. Запуск Pig


- Скрипт
    - Виконуються команди із файлу
    
    - `$pig script.pig`
    
- Grunt
    - Інтерактивна оболонка для виконання команд Pig

    - Запускається у разі відсутності скрипта

### 7. Pig Latin

- Будівельні блоки
    - *Field (поле)* – частина даних

    - *Tuple (кортеж)* – впорядкований набір полів, заключених у дужки "(" та ")"
        - Напр. `(10.4, 5, word, 4, field1)`
        
    - *Bag (мішок)* – колекція кортежів, заключених у дужки "{" та "}"
        - Напр. `{(10.4, 5, word, 4, field1), (this, 1, blah)}`
        
- Схожість із реляційними БД
    - Bag – це таблиця у БД
    
    - Tuple – це рядок у таблиці

    - Bag не вимагає, щоб усі tuples містили одне й те саме число полів (на відміну від реляційної таблиці)

### Операції DUMP та STORE

- Жодних дій доти, доки не зустрінуться команди `DUMP` або `STORE`
    - Pig буде парсити, перевіряти та аналізувати вирази
    
    - Але не виконуватиме їх

- `DUMP` – виводить результат на екран

- `STORE` – зберігає результат (зазвичай у файл)



### Великий обсяг даних

- Дані в Hadoop зазвичай великого обсягу і немає сенсу їх всіх виводити на екран

- Зазвичай дані знаходяться у *HDFS/Hbase/etc*
    - Команда `STORE`

- З метою налагодження можна виводити лише невелику частину на екран



### Команда LOAD

`LOAD 'data' [USING function] [AS schema];`

- `data` – ім'я директорії або файлу
    - Повинно бути в одинарних лапках

- `USING` – визначає функцію, що використовується для завантаження
    - За замовчуванням, використовується `PigStorage`, яка парсить кожний рядок, використовуючи роздільник
        - <span style="font-size:smaller;">За замовчуванням, розділювач знак табуляції ('\t')</span>
        - <span style="font-size:smaller;">Розділювач може бути заданий, використовуючи регулярні вирази</span>

- `AS` – визначає схему вхідних даних
    - призначає імена полям
    
    - оголошує тип полів

### PigLatin: FOREACH

- `FOREACH <bag> GENERATE <data>;`
    - Ітерація по кожному елементу в bag та його обробка

    - `grunt> result = FOREACH bag GENERATE f1;`
    


### PigLatin: FOREACH з функцією

- `FOREACH B GENERATE group, FUNCTION(A);`

    - Pig містить багато вбудованих функцій
        `COUNT`, `FLATTEN`, `CONCAT` і т.д.
        
    - Можна реалізувати свою функцію UDF (*User Defined Functions*)
        - Java, Python, JavaScript, Ruby або Groovy

### PigLatin: Joins

- Pig підтримує
    - Inner Joins
    
    - Outer Joins
    
    - Full Joins

- Фази join
    - Завантажити записи в bag з input \#1

    - Завантажити записи в bag з input \#2
    
    - Зробити join для двох масивів даних (bags) по заданому ключу

- За замовчуванням використовується **Inner Join**
    - Об'єднуються рядки за ключами
    
    - Рядки, які не мають відповідностей, не включаються до результату

### PigLatin: Outer Join

- Рядки, що не мають відповідностей в "іншій" таблиці, всерівно будуть включені до результату

- `JOIN <bag #1> BY <join field> LEFT OUTER, <bag #2> BY <join field>;`

- `JOIN <bag #1> BY <join field> RIGHT OUTER, <bag #2> BY <join field>;`

- `JOIN <bag #1> BY <join field> FULL OUTER, <bag #2> BY <join field>;`
